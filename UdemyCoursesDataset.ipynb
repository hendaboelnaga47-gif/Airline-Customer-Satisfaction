```python
# Temporarily convert the column to string to handle $ or "Free"
df['price'] = df['price'].astype(str) 

# Replace "Free" with 0 and remove $
df['price'] = df['price'].replace('Free', '0')
df['price'] = df['price'].str.replace('$', '', regex=False)

# Convert the column to numeric (float)
df['price'] = df['price'].astype(float)

# check
print(df['price'].head())
```

```python
def duration_to_minutes(duration):
    if pd.isnull(duration):
        return 0
    duration = str(duration).lower().strip()
    h, m = 0, 0
    
# Handle hours part
    if 'hour' in duration:
        h_part = duration.split('hour')[0].strip()
        try:
            h = float(h_part)  # بدل int نستخدم float
        except:
            h = 0
        duration = duration.split('hour')[1] if len(duration.split('hour')) > 1 else ''
    
    # Handle minutes part
    if 'min' in duration:
        m_part = duration.split('min')[0].strip()
        try:
            m = int(m_part)
        except:
            m = 0
    
    return int(h*60 + m)

# Recreate the column
df['content_duration_mins'] = df['content_duration'].apply(duration_to_minutes)

# check
print(df[['content_duration', 'content_duration_mins']].head(10))
```
